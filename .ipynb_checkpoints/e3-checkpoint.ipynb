```python
from __future__ import print_function, division

import thinkdsp as dsp
import matplotlib.pyplot as plt

%matplotlib inline
import seaborn

import numpy as np
import pandas as pd
import scipy.stats as stats
import statistics
import glob
```

# قسمت الف :

```python
framerate = 8000;

one = glob.glob("TrainSet\\1\\*.wav")
six = glob.glob("TrainSet\\6\\*.wav")
nine = glob.glob("TrainSet\\9\\*.wav")

one_powers = []
six_powers = []
nine_powers = []

for w in one: 
    wave = dsp.read_wave(w)
    wave_array = wave.ys
    wave_energy = np.sum(np.absolute(wave_array) ** 2)
    one_powers.append(wave_energy/len(wave_array))
    
for w in six: 
    wave = dsp.read_wave(w)
    wave_array = wave.ys
    wave_energy = np.sum(np.absolute(wave_array) ** 2)
    six_powers.append(wave_energy/len(wave_array))
    
for w in nine: 
    wave = dsp.read_wave(w)
    wave_array = wave.ys
    wave_energy = np.sum(np.absolute(wave_array) ** 2)
    nine_powers.append(wave_energy/len(wave_array))
    
```

```python
one_mu = statistics.fmean(one_powers)
one_sigma = statistics.stdev(one_powers, one_mu)

six_mu = statistics.fmean(six_powers)
six_sigma = statistics.stdev(six_powers, six_mu)

nine_mu = statistics.fmean(nine_powers)
nine_sigma = statistics.stdev(nine_powers, nine_mu)

one_dist_power = stats.norm(one_mu, one_sigma)
six_dist_power = stats.norm(six_mu, six_sigma)
nine_dist_power = stats.norm(nine_mu, nine_sigma)
```

```python
one_x_axis = np.arange(-0.05, 0.05, 0.001)
plt.plot(one_x_axis, one_dist_power.pdf(one_x_axis), linewidth=4)
plt.title("ONE's PDF [Power]")
plt.show()
plt.hist(one_powers, bins=50, color='c', density=True)
hist_min, hist_max = plt.xlim()
one_hist_x_axis = np.linspace(hist_min, hist_max, 200)
plt.plot(one_hist_x_axis, one_dist_power.pdf(one_hist_x_axis), linewidth=4)
plt.title("ONE's HISTOGRAM WITH PDF [Power]")
plt.show()
```

```python
six_x_axis = np.arange(-0.05, 0.05, 0.001)
plt.plot(six_x_axis, six_dist_power.pdf(six_x_axis), linewidth=4)
plt.title("SIX's PDF [Power]")
plt.show()
plt.hist(six_powers, bins=50, color='c', density=True)
hist_min, hist_max = plt.xlim()
six_hist_x_axis = np.linspace(hist_min, hist_max, 200)
plt.plot(six_hist_x_axis, six_dist_power.pdf(six_hist_x_axis), linewidth=4)
plt.title("SIX's HISTOGRAM WITH PDF [Power]")
plt.show()
```

```python
nine_x_axis = np.arange(-0.05, 0.05, 0.001)
plt.plot(nine_x_axis, nine_dist_power.pdf(nine_x_axis), linewidth=4)
plt.title("NINE's PDF [Power]")
plt.show()
plt.hist(nine_powers, bins=50, color='c', density=True)
hist_min, hist_max = plt.xlim()
nine_hist_x_axis = np.linspace(hist_min, hist_max, 200)
plt.plot(nine_hist_x_axis, nine_dist_power.pdf(nine_hist_x_axis), linewidth=4)
plt.title("NINE's HISTOGRAM WITH PDF [Power]")
plt.show()
```

```python
x_axis = np.arange(-0.02, 0.05, 0.001)
plt.plot(x_axis, one_dist_power.pdf(x_axis), color = 'r')
plt.plot(x_axis, nine_dist_power.pdf(x_axis), color = 'b')
plt.title("ONE (RED) & NINE (BLUE) Intersection [Power]")
plt.show()
```

# قسمت ب :

```python

one = glob.glob("TrainSet\\1\\*.wav")
six = glob.glob("TrainSet\\6\\*.wav")
nine = glob.glob("TrainSet\\9\\*.wav")

one_zero_crossing_rates = []
six_zero_crossing_rates = []
nine_zero_crossing_rates = []

for w in one: 
    wave = dsp.read_wave(w)
    wave_array = wave.ys
    zero_crossing_rate = len(np.nonzero(np.diff(wave_array > 0))[0])
    one_zero_crossing_rates.append(zero_crossing_rate)
    
for w in six: 
    wave = dsp.read_wave(w)
    wave_array = wave.ys
    zero_crossing_rate = len(np.nonzero(np.diff(wave_array > 0))[0])
    six_zero_crossing_rates.append(zero_crossing_rate)
    
for w in nine: 
    wave = dsp.read_wave(w)
    wave_array = wave.ys
    zero_crossing_rate = len(np.nonzero(np.diff(wave_array > 0))[0])
    nine_zero_crossing_rates.append(zero_crossing_rate)
    
```

```python
one_mu = statistics.fmean(one_zero_crossing_rates)
one_sigma = statistics.stdev(one_zero_crossing_rates, one_mu)

six_mu = statistics.fmean(six_zero_crossing_rates)
six_sigma = statistics.stdev(six_zero_crossing_rates, six_mu)

nine_mu = statistics.fmean(nine_zero_crossing_rates)
nine_sigma = statistics.stdev(nine_zero_crossing_rates, nine_mu)

one_dist_zc = stats.norm(one_mu, one_sigma)
six_dist_zc = stats.norm(six_mu, six_sigma)
nine_dist_zc = stats.norm(nine_mu, nine_sigma)
```

```python
one_x_axis = np.arange(-5000, 5000, 1)
plt.plot(one_x_axis, one_dist_zc.pdf(one_x_axis), linewidth=4)
plt.title("ONE's PDF [Zero Crossing]")
plt.show()
plt.hist(one_zero_crossing_rates, bins=50, color='c', density=True)
hist_min, hist_max = plt.xlim()
one_hist_x_axis = np.linspace(hist_min, hist_max, 200)
plt.plot(one_hist_x_axis, one_dist_zc.pdf(one_hist_x_axis), linewidth=4)
plt.title("ONE's HISTOGRAM WITH PDF [Zero Crossing]")
plt.show()
```

```python
six_x_axis = np.arange(-5000, 5000, 1)
plt.plot(six_x_axis, six_dist_zc.pdf(six_x_axis), linewidth=4)
plt.title("SIX's PDF [Zero Crossing]")
plt.show()
plt.hist(six_zero_crossing_rates, bins=50, color='c', density=True)
hist_min, hist_max = plt.xlim()
six_hist_x_axis = np.linspace(hist_min, hist_max, 200)
plt.plot(six_hist_x_axis, six_dist_zc.pdf(six_hist_x_axis), linewidth=4)
plt.title("SIX's HISTOGRAM WITH PDF [Zero Crossing]")
plt.show()
```

```python
nine_x_axis = np.arange(-5000, 5000, 1)
plt.plot(nine_x_axis, nine_dist_zc.pdf(nine_x_axis), linewidth=4)
plt.title("NINE's PDF [Zero Crossing]")
plt.show()
plt.hist(nine_zero_crossing_rates, bins=50, color='c', density=True)
hist_min, hist_max = plt.xlim()
nine_hist_x_axis = np.linspace(hist_min, hist_max, 200)
plt.plot(nine_hist_x_axis, nine_dist_zc.pdf(nine_hist_x_axis), linewidth=4)
plt.title("NINE's HISTOGRAM WITH PDF [Zero Crossing]")
plt.show()
```
