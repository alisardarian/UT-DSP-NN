```python
from __future__ import print_function, division

import thinkdsp as dsp
import thinkplot as plt

%matplotlib inline

import numpy as np
import pandas as pd
import scipy.signal as signal
```

# قسمت الف :

```python
framerate = 16000;

aa_wave = dsp.read_wave('aa.wav')
s_wave = dsp.read_wave('s.wav')
aseman_wave = dsp.read_wave('aseman.wav')

plt.subplot(1,5,1)
plt.config(title = "AA.wav")
aa_wave.plot()
plt.subplot(3,5,1)
plt.config(title = "S.wav")
s_wave.plot()
plt.subplot(5,5,1)
plt.config(title = "ASEMAN.wav")
aseman_wave.plot()
```

```python
aa_wave_array = aa_wave.ys
s_wave_array = s_wave.ys
aseman_wave_array = aseman_wave.ys

aa_wave_duration = len(aa_wave_array) / framerate
s_wave_duration = len(s_wave_array) / framerate
aseman_wave_duration = len(aseman_wave_array) / framerate

print ("duration of AA.wave =", aa_wave_duration, "s")
print ("duration of S.wave =", s_wave_duration, "s")
print ("duration of ASEMAN.wave =", aseman_wave_duration, "s")
```

```python
aa_wave_energy = np.sum(np.absolute(aa_wave_array) ** 2)
s_wave_energy = np.sum(np.absolute(s_wave_array) ** 2)
aseman_wave_energy = np.sum(np.absolute(aseman_wave_array) ** 2)

print ("energy of AA.wave =", aa_wave_energy)
print ("enegy of S.wave =", s_wave_energy)
print ("energy of ASEMAN.wave =", aseman_wave_energy)
```

```python
print ("power of AA.wave =", aa_wave_energy/len(aa_wave_array))
print ("power of S.wave =", s_wave_energy/len(s_wave_array))
print ("power of ASEMAN.wave =", aseman_wave_energy/len(aseman_wave_array))
```

# قسمت ب :

```python
aa_wave_filtered_100 = dsp.Wave(signal.medfilt(aa_wave_array, 101), None, framerate)
aa_wave_filtered_1000 = dsp.Wave(signal.medfilt(aa_wave_array, 1001), None, framerate)
aa_wave_filtered_3000 = dsp.Wave(signal.medfilt(aa_wave_array, 3001), None, framerate)

plt.subplot(1,7,1)
plt.config(title = "AA.wav (normal)")
aa_wave.plot()
plt.subplot(3,7,1)
plt.config(title = "AA.wav (filtered with N=100)")
aa_wave_filtered_100.plot()
plt.subplot(5,7,1)
plt.config(title = "AA.wav (filtered with N=1000)")
aa_wave_filtered_1000.plot()
plt.subplot(7,7,1)
plt.config(title = "AA.wav (filtered with N=3000)")
aa_wave_filtered_3000.plot()
```

```python
aa_wave_filtered_100.write('aa_filtered_100.wav')
aa_wave_filtered_100.make_audio()
```

# همچنین مقدار نویز در سیگنال س بیشتر از دو سیگنال دیگر است که می‌تواند به دلیل بالا بودن شدت تغییرات سیگنال س نسبت به دو سیگنال دیگر باشد و مقدار بیشتری از نمونه‌ها در سیگنال س از بین می‌روند

# قسمت ج :

```python
aa_wave_fft = np.fft.fft(aa_wave_array)
plt.plot(np.absolute(aa_wave_fft))
print ("first 4 elements :\n", aa_wave_fft[:4])
print ("last 4 elements :\n", aa_wave_fft[-4:])
```

```python
plt.plot(np.angle(aa_wave_fft, deg = False))
```

```python
s_wave_fft = np.fft.fft(s_wave_array)
plt.plot(np.absolute(s_wave_fft))
print ("first 4 elements :\n", s_wave_fft[:4])
print ("last 4 elements :\n", s_wave_fft[-4:])
```

```python
plt.plot(np.angle(s_wave_fft, deg = False))
```

# مشاهده می‌شود که به غیر از عنصر اول 4 نمونه‌ی اول و آخر طیف سیگنال‌ها با هم برابر بوده و نمودار طیف نیز حالت تقارن دارد
# نمودار فاز دو سیگنال تقریبا تفاوتی با هم ندارند و تفاوت میان فاز‌های دو سیگنال کم است
